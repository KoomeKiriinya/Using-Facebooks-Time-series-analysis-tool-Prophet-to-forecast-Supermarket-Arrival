```python
# import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
from matplotlib.pylab import rcParams
from fbprophet import Prophet
from datetime import datetime
from fbprophet.diagnostics import cross_validation
rcParams['figure.figsize'] = 18, 8
```

```python
# get the csv
retailData = pd.read_csv("retail_dataset2.csv")
copyData = retailData.copy()
copyData['Datetime']=pd.to_datetime(copyData['trans_date']).apply(lambda x : str(x.day) +str("/") +str(x.month) +str('/')+ str(x.year))
```

```python
# remove all empty columns
copyData=copyData.dropna()
# get all dates
copyData['Datetime']= pd.to_datetime(copyData['Datetime'])
Times = list(copyData['Datetime'].unique())
```

```python
# calculating the number of customers based on the receipt number
noCust=[]
for i in Times:
    cust = len(list(copyData['receipt_no'][copyData['Datetime']==i].unique()))
    noCust.append(cust)
```

```python
# create a new data frame to be used for analysis
newDataSet = pd.DataFrame()
newDataSet['ds'] = pd.to_datetime(Times)
newDataSet['y']= noCust
newDataSet= newDataSet.sort_values('ds')
# our dataset contains a column with the date and the total number of receipts(total number of receipts = total number of customers)
newDataSet.head()
```

# Plotting our Historical Data

```python
#removing certain parts of the date to remove bias
newDataSet=newDataSet.iloc[2:2100,:]
# There is a relatively high variance so we take the log of they y 
newDataSet['y']=np.log(newDataSet['y'])
plt.plot(newDataSet['ds'],newDataSet['y'])
# add required components to the plotting
plt.xlabel("Year",fontsize=20)
plt.ylabel("Number of customers",fontsize=20)
plt.title("Customer arrival Trend",fontsize=20)
plt.show()
```

# Fitting our historical data and making predictions

```python
# fitting our model
m = Prophet()
m.fit(newDataSet)
# predict the future using our model
future=m.make_future_dataframe(freq='D',periods = 1*365 ,include_history=False)
pred=m.predict(future)
pred[['ds','yhat','yhat_lower','yhat_upper']].tail()

```

# Forecasted Customer Trend ie. general trend, yearly trend, weekly trend

```python
fig = m.plot_components(pred)
```

# From the graph we observe
1. There is a general downward trend of customers visiting with loyalty cards.
2. There is a exponential increase of customers visiting with loyalty cards towards the end of the year
3. Thers is an increase in customers visiting with loyalty card towards the end of the week
